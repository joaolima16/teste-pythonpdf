```python
import pdfplumber
def extractText():
    textPDF = ""
    with pdfplumber.open("Jeruel.pdf") as archive:
        for doc in archive.pages:
            textPDF += doc.extract_text()
    return textPDF


extractText()
```

```python
import pdfplumber
import spacy

nlp = spacy.load("pt_core_news_sm")
text = extractText()

doc = nlp(text)

for ent in doc:
    numeros = [ent.text for ent in doc if ent.like_num]
 
    # print(numeros)
    print(ent)
```
